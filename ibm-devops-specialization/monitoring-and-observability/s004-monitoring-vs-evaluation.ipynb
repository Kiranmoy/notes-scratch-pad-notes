# Monitoring and evaluation



Monitoring and evaluation can often be confused with one another.

Yet, there are some significant differences.

Monitoring is a routine, ongoing process:
* It is an operational-level activity.
* It’s usually performed by developers and their teams.

In contrast, evaluations are long-term processes that may happen only a few times.
* Evaluation is a business-level activity.
* It is usually performed by others not involved in developing the application.

# Monitoring process



* When we talk about monitoring, we’re talking about the process performed by a team that is directly involved with the application and services.
* Monitoring is used to collect data at multiple points in the programming, testing, and post-deployment lifecycle.

# Monitoring benefits



* Monitoring is done regularly and systematically.
* Developers use monitoring to track application performance.
* Through monitoring, application management tools can identify, measure, and evaluate how well an application is working.
* Also, monitoring allows developers to collect and measure data, which can help visualize any issues or unexpected events that may occur while an application is running.
* Monitoring also gathers essential information, such as application availability, occurrences of issues and bugs, resource usage, and changes that might alter the user experience.
* This information can be used for evaluation.

# Evaluation process



* Unlike monitoring, evaluations can be performed by others who may manage the app but did not develop the application, like managers, supervisors, or external independent parties.
* Impartial evaluations look at the overall value of an application.
* Evaluations might be done during or at the end of a rollout of an application.

# Evaluation benefits



* Additionally, evaluation is used to assess whether a solution meets the goals identified at the design stage or when the solution was implemented.
* It also helps determine if the solution has resulted in anticipated outcomes or if anything unanticipated has occurred.
* Evaluation stresses improving the effectiveness of the project and meeting established standards.

# Monitoring and evaluation



Monitoring is important because it helps developers keep an application running.
* Identifies ways to improve the application’s performance.
* Importantly, monitoring may expose any possible security issues.

Evaluation is also important because it shows how well an app performs and if it does what it was designed to do.
* Evaluation can also show how useful an app is.
* How sustainable and profitable it is.

# Why two different processes? 



The two separate processes exist because they do different things: 
* Monitoring helps developers see that their application is performing as it should.
* Evaluation shows that an application is both aligned with outcomes and is profitable.
* When combined, these two processes ensure that a healthy, valuable, and performant application is delivered to the users.

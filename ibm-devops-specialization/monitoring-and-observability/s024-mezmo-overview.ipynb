# Mezmo and its use cases



* Mezmo, previously known as LogDNA, is a robust observability platform for managing and taking action on your telemetry data.
* Mezmo log analysis gives operations teams control over their data flow and empowers developers to find value from their logs quickly.
* You can easily centralize your logs from multiple sources with minimal setup and onboarding using the Mezmo agents, code libraries, or APIs.

# Features of Mezmo



Working with Mezmo offers several features.

Mezmo was built with the entire DevOps team in mind.
* You can configure the features so that operations teams can get the correct data to the correct development team faster, resulting in more secure applications while increasing collaboration.
* It can be used for automatic and custom parsing.
* It allows developers to easily surface the data they need from their logs.

You can leverage Mezmo's alert feature to get notifications about your system's activity.
* You can also set up alerts to notify your team of issues occurring in production.
* Alerts could be configured using several integrations such as PagerDuty, Slack, and more.

Mezmo gives you the ability to see critical Cloud and Kubernetes events.

You can visualize your log data over a time period to better analyze your system's activity and identify trends.

With spike protection features, you can set dynamic thresholds and alerts when data volume limits are hit.

Mezmo complies with industry regulations such as:
* Payment Card Industry (PCI) 
* Service Organization Control type 2 (SOC 2) 
* the Health Insurance Portability and Accountability Act (HIPAA)
* the General Data Protection Regulation, or GDPR)

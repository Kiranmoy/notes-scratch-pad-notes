# What is application monitoring?



* Application monitoring is a process developers use to ensure their software performs as intended.
* Developers use monitoring tools to gather performance metrics to evaluate how an application functions.
* It allows rapid detection of any issues, bugs, or unexpected events that occur in real-time while the application is running.
* Application monitoring provides a better understanding of app usage, allowing developers to fine-tune and maintain apps for increased performance and user satisfaction.

# Functions of application monitoring



There are several functions of application monitoring tools.
* Observation of components, like servers, databases, message queues, and caches, is essential to ensure that your application and its dependencies are running smoothly and as intended.
* Monitoring tools also provide visualization and alerts.
* Dashboards provide an overview, and alerts call attention to specific problems.
* Additionally, monitoring tools can detect anomalies, varying from simple threshold detection to advanced machine learning pattern recognition.
* With distributed tracing, developers can monitor how one event connects across multiple nodes to detect the origins of errors.
* Another main function of monitoring tools is dependency mapping, which visually represents how requests travel between services.


# How application monitoring works?



* Application monitoring is typically delivered on-premises or from cloud-based solutions.
*  It can also be a software-based or hardware-based solution.
* Hardware-based solutions include specialized web appliances that gather performance telemetry, which is the data and metrics gathered and recorded by your system.
* Software-based agent processes can be deployed alongside a web application.
* These agents hook into application performance indicator (or API) runtimes to collect telemetry data.
* This works with synthetic traffic to augment monitoring.
* Synthetic traffic simulates user interaction to monitor application throughput and performance at designated intervals.
* Typically, it is generated by an external application installed on various endpoints, which can be located anywhere in the world.

# Telemetry monitored



* Telemetry is system data that is automatically gathered and recorded for monitoring.
* Telemetry includes resource utilization, server logs, network traffic, and performance metrics.
* Developers and their teams monitor: 
    * CPU and other resource utilization metrics from the servers hosting and running your applications.
    * Server logs to compare the number of user requests received versus response time and tracking error rates on target applications.
    * Network traffic to detect application downtime.
    * Key performance metrics, which are measured and analyzed against application dependencies on programming languages and services.

# What are some of the benefits of application monitoring? 



* Rapid diagnosis is one key benefit.
* The immediate benefit of application monitoring is proactive troubleshooting – in other words, the ability to rapidly diagnose and fix future issues before they impact end users.
* Also, application monitoring can deliver: faster resolution times, and improved IT staff productivity.
* When used consistently as part of DevOps processes, application monitoring can help: accelerate application deployment, and help you deliver higher-quality user experiences.

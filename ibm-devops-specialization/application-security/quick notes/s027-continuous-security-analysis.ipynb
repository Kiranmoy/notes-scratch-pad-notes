# Continuous security analysis



* Continuous security analysis is the ongoing integration and testing of security within the software development lifecycle (or SDLC).
* Starting with Requirements, you conduct continuous security analysis through the Design stage, the Develop stage, during Testing, and when you deploy the code.
* Continuous security testing and analysis should be a priority from the initial stage of development and beyond deployment to when the code is in production.

# Goals



Next, **what are the key goals of continuous security analysis?** 
* Continuous security analysis reduces security risks, and it helps you remove potential bottlenecks in development.
* So, you want to conduct continuous security analysis to deliver fast software releases.

# Secure SDLC



* By implementing continuous security analysis, you will create a secure SDLC for your organization.
* When you implement continuous security analysis, security becomes a priority and a critical factor in the SDLC.

# Benefits



So, integrating continuous security analysis in the SDLC provides many benefits.
* First, you can integrate continuous security analysis directly into a continuous integration/continuous delivery (or CI/CD) pipeline.
* Next, because continuous security analysis is a high priority throughout the SDLC, you can fix security issues early in the development process.
* Continuous security analysis accelerates new feature development because when you ensure security, you reduce disruption in development.
* Another benefit of continuous security analysis is that you can automate security requirements for existing and new application development.
* Continuous security analysis improves software ownership and responsibility among developers.
Secure coding becomes a default developmental style.

# Snyk Open Source tool



* To perform continuous security analysis, one tool you can use is Snyk Open Source.
* This tool finds and fixes vulnerabilities in open source libraries.
* It finds licensing issues caused by open source libraries.
* It offers advice for fixing vulnerabilities during development.
* It fixes vulnerabilities using pull requests when you are developing with a source control management (or SCM) coding repository such as GitHub.
* Snyk Open Source prioritizes issues and reports on newly discovered vulnerabilities.

# Best practices



To implement continuous security analysis, follow these four best practices.
* Require planning to audit security issues and risks throughout development.
* Develop your code on up-to-date operating systems with software in a secure environment.
* Analyze all components of code development to build security directly into the full development stack.
* Implement static application security testing (or SAST) and software composition analysis (or SCA) in addition to continuous security analysis to ensure that all code testing meets multiple valid testing criteria.

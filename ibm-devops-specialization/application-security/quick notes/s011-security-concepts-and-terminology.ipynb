# Authentication & Authorization

First, you hear about authentication and about authorization.

**What’s the difference?**
* Well, authentication is the process of verifying a user’s identity.
* It means ensuring someone really is who they say they are.
* Their identity is authentic.

On the other hand, **authorization** is the process of determining a user’s access rights.

**In other words, now that I know who you are; what are you allowed to do?**
* So, once you authenticate a user, they can become authorized.
* Authorization ensures that they can do only what they are allowed to do.

# Encryption

Next, **what exactly is encryption?** 
* Encryption is the process of encoding information so that only those users with authorized access can decode it.
* Authorized access means they have the correct access rights.
* Encryption has two main types.
    * **Symmetric encryption** is when the same key is used for both encrypting and decrypting.
    * **Asymmetric encryption** is when different keys are used to encrypt and decrypt.

# Integrity

Next, integrity is one of those terms that seems to have no obvious exact definition.
In fact, it does! 
* Integrity is the process of ensuring that data hasn’t been changed by an unauthenticated source.
* It means you can trust the data that you are reading.
* For instance, one way to achieve integrity is to use secure hash algorithms.
* This creates a hash of the data, so that you can check if applying the algorithm later leads to a different result than expected, which would indicate that the data has been tampered with.

# CI/CD Pipeline



Another term you may have come across is CI/CD.

The main idea of CI/CD is to use automation in the application development pipeline.

The CI part of CI/CD stands for continuous integration, which means that new code changes are regularly built, tested, and merged.

The CD part of the CI/CD actually stands for two concepts: 
* The first, continuous delivery, which means that tested changes are automatically released to a development, test, or staging environment; 
* The second, continuous deployment, which means that the new releases are automatically deployed to production.

# Security in CI/CD and the SDLC

Security in the CI/CD pipeline is vital to the software development lifecycle (or SDLC).

You can apply security with relative ease.



For example: 
* Scan source code early in the continuous integration stage.
* Perform source code scanning and analysis to expose vulnerabilities.
* Then, add more tests in the CI/CD pipeline.
* Perform security tests along with other CI/CD tests.
* Even after deployment, continuously detecting and report new threats.
* Integrate runtime security to detect threats that arise in production.

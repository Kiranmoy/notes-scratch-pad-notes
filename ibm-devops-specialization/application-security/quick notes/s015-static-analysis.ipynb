# What is static analysis?

* Static analysis examines all code or runtime binaries to help detect common vulnerabilities.
* It's a debugging method that automatically inspects source code before execution.

# Static application security test

* **Static application security testing** (or **SAST**) examines source code to identify security flaws that render your organization's applications vulnerable to attack.
* It is extremely effective in detecting problems in code that you evaluated manually.
* Some SAST solutions provide application programming interfaces (or APIs) for DevOps integration and do not require code to be complete.
* Static analysis can take a long time because it thoroughly scans the code.

# Static analysis in SDLC



**Where does static analysis belong in the software development lifecycle (or SDLC)?** 
* Static code analysis takes place early in the development process before software testing begins.
* For DevOps enterprises, static analysis occurs during the **Develop** stage and establishes an automatic feedback loop.
* So, you will become aware of any issues with your code from the start.
* It will be simpler for you to resolve those issues.

# Static analysis tools

* Tools for static analysis examine your code statically on the file system.
* It’s referred to as static because the code is not executing.

# Benefits of static analysis

Here are three of the most significant benefits of using static analysis tools.

The best static code analysis tools offer **depth**, **speed**, and **accuracy**.

First, static analysis provides **depth**.
* Although testing cannot cover every conceivable avenue of code execution, the static code analyzer can cover most of them.
* As you work on your build, static analysis verifies the code.
* Based on the criteria that you've set, you will receive an in-depth analysis of where possible problems may exist in your code.

Next, static analysis provides **speed**, which helps you deliver high-quality software quickly and at an affordable price.
* Automated tools are substantially faster than manual analyzers.
* Static code inspection identifies issues early on.
* It pinpoints the exact location of the problem in the code.
* As a result, you will be able to correct issues more quickly.
* Early code problems are less expensive to fix.

**Accuracy** is another benefit.
* Automated static analysis tools provide a high degree of accuracy.
* Manual reviews require a high degree of knowledge about vulnerabilities, while automated scanning can leverage more in-depth knowledge of the latest vulnerabilities.
* Automated tools examine every line of code for potential errors.
* When you know what those are, you can put the best code in place before testing begins.
* After all, when it comes to coding standards, quality is quintessential.
